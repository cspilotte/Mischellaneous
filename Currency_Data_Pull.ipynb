```python
import numpy as np
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
import time
```

```python
def currency_pull(date):
    time.sleep(5)
    url = 'https://www.imf.org/external/np/fin/data/rms_mth.aspx?SelectDate={}&reportType=CVSDR'.format(date)
    html = urlopen(url)
    soup = BeautifulSoup(html)
    soup_tables = soup.findAll('table', class_='tighter')
    headers = [th.getText().replace('\r\n', '').replace(' ', '') 
           for th in soup_tables[0].findAll('th', class_='color3')]
    values = soup_tables[0].findAll('td', align='right')
    headers1 = [th.getText().replace('\r\n', '').replace(' ', '') 
            for th in soup_tables[1].findAll('th', class_='color3')]
    values1 = soup_tables[1].findAll('td', align='right')
    rows = [td.getText().replace('\r\n', '').replace(' ', '') 
        for td in soup_tables[0].findAll('td', align = 'left')]
    df = pd.DataFrame(index = rows, columns = headers[1:])
    count = 0
    for i in range(0, len(df.index)):
        for j in range(0, len(df.columns)):
            df.iloc[i, j] = values[count+j].getText().replace('\r\n', '').replace(' ', '').replace(',', '')
        #print(values[i].getText().replace('\r\n', '').replace(' ', ''))
        count += len(df.columns)
    df1 = pd.DataFrame(index = rows, columns = headers1[1:])
    count = 0
    for i in range(0, len(df1.index)):
        for j in range(0, len(df1.columns)):
            df1.iloc[i, j] = values1[count+j].getText().replace('\r\n', '').replace(' ', '').replace(',', '')
        #print(values[i].getText().replace('\r\n', '').replace(' ', ''))
        count += len(df1.columns)
    df = pd.concat([df, df1], axis = 'columns')
    return df
```

```python
currency_pull('2010-03-31')
```

```python
month_day = ['01-31', '02-28', '03-31', '04-30', '05-31', '06-30', '07-31', '08-31', '09-30', '10-31', '11-30', '12-31']
```

```python
df = currency_pull('2020-12-31')
for i in range(2021, 2023):
    for j in month_day:
        date = str(i) + '-' + j
        print(date)
        df = pd.concat([df, currency_pull(date)], axis = 'columns')
df = df.replace('NA', np.nan)
df
```

```python
#we now will impute the NaN values. We can use 
from sklearn.impute import SimpleImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
#transpose the df to put it in time series format
df_ts = df.transpose()
df_ts.head()
```

```python
#We can now impute the missing values using the simple imputer
#We will use the mean
mean_imputer = SimpleImputer(strategy='mean')
mean_imputer = mean_imputer.fit(df_ts)
df_mean_impute = pd.DataFrame(mean_imputer.transform(df_ts), columns = df_ts.columns, index = df_ts.index)
df_mean_impute.head()
```

```python
#We can also use the median
med_imputer = SimpleImputer(strategy='median')
med_imputer = med_imputer.fit(df_ts)
df_med_impute = pd.DataFrame(med_imputer.transform(df_ts), columns = df_ts.columns, index = df_ts.index)
df_med_impute.head()
```

```python
#The iterative imputer offers a more sophisticated method of imputation
imputer = IterativeImputer(random_state = 42)
imputed_df = pd.DataFrame(imputer.fit_transform(df_ts), columns = df_ts.columns, index = df_ts.index)
imputed_df
```
