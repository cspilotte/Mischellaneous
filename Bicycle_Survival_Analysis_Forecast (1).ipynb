```python
#We want to build a predictive model that forecasts the number of loops done in central park on a 
#give day in the future. I will show how hazard modeling can be used to create a more accurate forecast
#when our training dataset is small (90 days)
```

```python
#For survival analysis we will need to reformat our data. First I will add in an initial date
#We will use the start of Quarter 2 for this
df_survival = sqldf('''select *
                       from (select bikeid, date, loop_count
                             from df_loop_counts
                             union 
                             select distinct bikeid, "2016-04-01" as date, 0 as loop_count
                             from df_loop_counts 
                             union
                             select distinct bikeid, "2016-06-30" as date, 0 as loop_count
                             from df_loop_counts)
                       order by bikeid, date''')
df_survival
```

```python
#Now we need to calculate the time between events for each of these bikers
#We can then use kaplan meyer estimation and hazard modeling to predict each bikers next event
df_time_event = sqldf('''select bikeid, 
                                lag(date, 1) over (PARTITION BY bikeid ORDER BY date) as first_date,
                                date as second_date,
                                JULIANDAY(date)-JULIANDAY(LAG(date, 1) OVER (PARTITION BY bikeid ORDER BY date)) as date_diff, 
                                loop_count
                         from df_survival
                         order by bikeid, date asc
                         ''')
df_time_event
```

```python
#we will also add in the rest of the bikers that traveled through Central Park & 6th 
#as people who "survived" the event of doing a loop during Q2
df_survive_pats = sqldf('''select distinct bikeid, "2016-04-01" as first_date, 
                                  "2016-06-30" as second_date, 
                                  julianday("2016-06-30")-julianday("2016-04-01") as date_diff,
                                  0 as loop_count
                           from df_cent_park
                           where bikeid not in (select distinct bikeid from df_time_event)
                           ''')
df_survival_final = sqldf('''select *
                             from df_time_event
                             union
                             select *
                             from df_survive_pats''')
df_survival_final
```

```python
#We can now begin our survival analysis
#We will consider instances of bikers who have their second date as 2016-06-30 as bikers who
#survived performing a loop. Other second dates will be instances where a biker does not survive
#performing the loop
df_survival_final1 = sqldf('''select distinct bikeid, date_diff, 
                                    case when "second_date" = "2016-06-30" then True
                                         else False end as Survive
                             from df_survival_final
                             where date_diff is not null''')
df_survival_final1
```

```python
#drop the bikeid to get our y_data for our survival analysis
y_surv_data = df_survival_final1.drop('bikeid', axis = 'columns')
y_surv_data.head()
```

```python
import lifelines
from lifelines import KaplanMeierFitter
```

```python
kmf = KaplanMeierFitter()
kmf.fit(durations = y_surv_data.date_diff, event_observed = y_surv_data.Survive)
kmf.plot_survival_function()
```

```python
from lifelines.utils import median_survival_times
median_ = kmf.median_survival_time_
median_confidence_interval_ = median_survival_times(kmf.confidence_interval_)
print(median_)
print(median_confidence_interval_)
```

```python
#we now must join our x_data in with our times and events
df_survival_all = sqldf('''select a.*, gender, usertype, total_trip, num_rides, age
                           from df_survival_final a
                           join df_survival_x b on a.bikeid=b.bikeid and a.first_date=b.date
                           where a.bikeid in (select distinct bikeid from df_time_event)
                           union
                           select a.*, gender, usertype, total_trip, num_rides, age
                           from df_survival_final a
                           join (select distinct bikeid, max(gender) as gender, max(usertype) as usertype, 
                                 max(total_trip) as total_trip, max(num_rides) as num_rides, max(age) as age
                                 from df_survival_x) b on a.bikeid=b.bikeid
                           where a.bikeid not in (select distinct bikeid from df_time_event)''')
df_survival_all
```

```python
#drop bikeid, first_date, and second_date
#
df_survival_final_all = sqldf('''select distinct bikeid, date_diff, 
                                    case when "second_date" = "2016-06-30" then True
                                         else False end as Survive,
                                    gender, usertype, total_trip, num_rides, age
                             from df_survival_all
                             where date_diff is not null''')
df_survival_final_all
surv_data_all = df_survival_final_all.drop('bikeid', axis = 'columns')
```

```python
from lifelines import CoxPHFitter
```

```python
#We can now perform our hazard modeling
cph = CoxPHFitter()
cph.fit(surv_data_all, duration_col = 'date_diff', event_col = 'Survive')
cph.print_summary()
```

```python
#we can use predict_median to get the number of days before the next loop each biker will perform
X = surv_data_all.drop(['Survive'], axis = 'columns')
cph.predict_median(X)
```

```python
#for our input data, we need to make a dataframe looking from the current day back
df_current_pre = sqldf('''select distinct a.bikeid, index_date, tripduration, 1 as ride, date,
                          case when usertype = "Subscriber" then 1 else 0 end as usertype
                      from (select distinct bikeid, max(date) as index_date
                            from df_loop_counts group by 1) a
                      join df_cent_park b on a.bikeid = b.bikeid
                      where b.date >= a.index_date
                      group by 1''')
df_current = sqldf('''select distinct bikeid, julianday('2016-07-01')-julianday(index_date) as date_diff,
                             sum(tripduration) over (partition by bikeid order by date) as total_trip,
                             sum(ride) over (partition by bikeid order by date) as num_rides,
                             sum(usertype) over (partition by bikeid order by date) as usertype
                      from df_current_pre''')
df_current_final = sqldf('''select distinct a.*, gender, age
                            from df_current a
                            join df_survival_x b on a.bikeid = b.bikeid''')
df_current_final
```

```python
#we now use our survival function to calculate the probability of survival over time for each bike
surv_matrix = cph.predict_survival_function(df_current_final)
#if we subtract each of these probabilities from 1 we get the probability that a bike will do
#a loop on a given day
loop_matrix = 1 - surv_matrix
loop_matrix
```

```python
#we can now compute expectation of a loop for each of these r.v's 
#the mle is the mean number of loops performed on days when a bike performs a loop 
#as the value of x associated with these probabilities.
mean_loops = np.mean(df_loop_counts.loop_count)
expected_loops = []
for i in range(0, len(loop_matrix.index)):
    x = mean_loops*loop_matrix.iloc[i,:]
    expected_loops.append(np.sum(x))
expected_loops = pd.Series(expected_loops, name = 'expected number of loops')
expected_loops
```

```python
#we can now difference this series to get the number of expected loops on a given day
loop_forecast = expected_loops.diff().dropna()
```
